```python

#!pip install Lark
from lark import Lark,Token,Discard
from lark import Transformer

class TransformaLista(Transformer):
    count = 0    
    avgsum = 0
    avgcount = 0

    def start(self,item):
        # [1] para trabalhar so com a lista de elementos sem os brakets redundantes
        print("nr alunos: " + str(self.count))
        return item[1]
    

    def alunos(self, elemento):
        #print(elemento)
        return elemento
    
    def PO(self,pe):
        return Discard

    def PC(self,pd):
        return Discard

    def VIR(self,vir):
        return Discard
    
    def SPACE(self,spc):
        return Discard
    
    def DOT(self,dot):
        return Discard
    
    def NL(self,nl):
        return Discard

    def NUMBER(self,num):
        self.avgcount +=1
        self.avgsum += int(num)
        return int(num)
    
    def VNUMBER(self,num):
        #print(num[1])
        return int(num[1])
    
    def alunonl(self,al):
        return al[0]
    
    def aluno(self,elem):
        self.count +=1
        #nome = elem
        l = elem[1]
        #print(l)
        #avg = sum(l)/len(l)
        print("avg: " + str(self.avgsum/self.avgcount))
        #print(lista)
        return l
        
    def lista(self,elems):
        #print(elems[0])
        return elems[0]
    
    def elems(self,elem):
        return elem
    
    def nome(self,elem):
        self.avgcount = 0
        self.avgsum = 0
        return elem


grammar = r'''
start : "TURMA" /[A-Z]\n/ alunos DOT
alunos: alunonl* aluno
alunonl : aluno NL
aluno : nome lista
lista : PO elems PC
elems : NUMBER (VNUMBER)*
nome : WORD
VNUMBER : VIR NUMBER
SPACE : " "
NL : ";"
PO : "("
PC : ")"
VIR : ","
DOT : "."
NUMBER : /[0-9]+/
%import common.WS
%import common.WORD
%ignore WS
'''

l = Lark(grammar)
frase = '''TURMA A
bro (2,3,4,5);
foo (5,4,3).
'''
tree = l.parse(frase)

data = TransformaLista().transform(tree)
print(data)
```
